```python
import pandas as pd

data = pd.read_csv (r'yellow_tripdata_2021-01.csv') 
print (data)
```

```python
#print(data["tpep_pickup_datetime"])
#print(data["tpep_dropoff_datetime"])

data["pickuptime"] = pd.to_datetime(data["tpep_pickup_datetime"])
print(data["pickuptime"])
data["droptime"] = pd.to_datetime(data["tpep_dropoff_datetime"])
print(data["droptime"])

print(data)
```

```python
data["difference_in_time"] = data["droptime"] - data["pickuptime"]
print(data["difference_in_time"])
```

```python
import matplotlib.pyplot as plt

plt.bar(data["total_amount"],data["trip_distance"])
plt.title('Distance V/s Amount')
plt.xlabel('total_amount')
plt.ylabel('trip_distance')
plt.show()
```
